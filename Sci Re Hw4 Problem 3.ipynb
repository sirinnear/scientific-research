```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
def cancer_sim(n, br, r):
    cancer_count = 0
    for i in range(n):
        if (np.random.random() < br+r): cancer_count += 1
    return cancer_count

cancer_sim(100, 0.4, 0.1)
```

```python
N_samples = [10**1, 10**2, 10**3, 10**4, 10**5, 10**6]
r_rate = [0, 0.001, 0.01, 0.1, 0.3]
```

```python
fig1, axes = plt.subplots(2,3, figsize=(25,10))

n = 0
m = 0
for i in range(6):
#     print('N = %i for 10 iterations' %N_samples[i])
    res = [cancer_sim(N_samples[i], 0.4, 0) for k in range(10)]
    axes[n, m].hist(res, label='N = %i' %N_samples[i], bins=None)
#     print(res)
#     print()
    
    if (m >= 2):
        m = 0
        n += 1
    else: m += 1
```

```python
fig2, axes = plt.subplots(2, 3, figsize=(25,10))

n = 0
m = 0
for i in range(6):
    print('N = %i for 10 iterations' %N_samples[i])
    for j in range(5):
        res = [cancer_sim(N_samples[i], 0.4, r_rate[j]) for k in range(10)]
        axes[n, m].hist(res, label='r = %f' %r_rate[j], bins=None)
        print(res)
    print()
    if (m >= 2):
        m = 0
        n += 1
    else: m += 1
    
for i, axes in enumerate(axes.flatten()):
    axes.set_title('N = %i' %N_samples[i])
    axes.set(xlabel='N = %i' %N_samples[i], ylabel='Number of iterations')
    
axes.legend()
```
