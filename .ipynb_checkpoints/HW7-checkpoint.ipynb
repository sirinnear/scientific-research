```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import math
import random
```

```python
n = 10000
x = np.array([np.random.normal(10, 0.4) for i in range(n)])
```

```python
plt.hist(x, bins=100);
plt.axvline(10, color='r')
plt.axvline(10+0.4, color='pink')
plt.axvline(10-0.4, color='pink')

print("X mean = 10")
print("X standard deviation = 0.4")
```

```python
n = 10000
y = np.array([np.random.normal(6, 0.3) for i in range(n)])
```

```python
plt.hist(y, bins=100);
plt.axvline(6, color='r')
plt.axvline(6+0.3, color='pink')
plt.axvline(6-0.3, color='pink')

print("Y mean = 6")
print("Y standard deviation = 0.3")
```

```python
z = x + y

mean = np.mean(z)

std = np.std(z)

fig1, (x1, y1, z1) = plt.subplots(3, 1, figsize=(5, 15))

x1.hist(x, bins=100);
x1.axvline(10, color='r')
x1.axvline(10+0.4, color='pink')
x1.axvline(10-0.4, color='pink')
x1.set_title("X");

y1.hist(y, bins=100);
y1.axvline(6, color='r')
y1.axvline(6+0.3, color='pink')
y1.axvline(6-0.3, color='pink')
y1.set_title("Y");

z1.hist(z, bins=100)
z1.axvline(mean, color='r')
z1.axvline(mean+std, color='pink')
z1.axvline(mean-std, color='pink')
z1.set_title("Z");

print("Z's mean =", mean)
print("Z's standard deviation =", std)
```

Z's mean and standard deviation looks similar to X's and Y's, despite Zz having a larger standard deviation.

```python
z = x - y

mean = np.mean(z)

std = np.std(z)

fig2, (x2, y2, z2) = plt.subplots(3, 1, figsize=(5, 15))

x2.hist(x, bins=100);
x2.axvline(10, color='r')
x2.axvline(10+0.4, color='pink')
x2.axvline(10-0.4, color='pink')
x2.set_title("X");

y2.hist(y, bins=100);
y2.axvline(6, color='r')
y2.axvline(6+0.3, color='pink')
y2.axvline(6-0.3, color='pink')
y2.set_title("Y");

z2.hist(z, bins=100)
z2.axvline(mean, color='r')
z2.axvline(mean+std, color='pink')
z2.axvline(mean-std, color='pink')
z2.set_title("Z");

print("Z's mean =", mean)
print("Z's standard deviation =", std)
```

Z's mean and standard deviation looks similar to X's and Y's, despite Z having a larger standard deviation.
