```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import math
import random
```

```python
fig1, plots = plt.subplots(1, 3, figsize=(20,5))

def generate_baby(avg_days, n):
    days = 0
    hist = []
    for i in range(1000):
        boy = 0
        for j in range(avg_days):
            if random.randint(0,1) == 0:
                boy += 1
        prob = boy/avg_days
        hist.append(prob)
        if prob >= 0.6:
            days += 1
    plots[n].hist(hist, bins=10)
    plots[n].set_ylim([0,500])
#     plots[n].set_xlim([0.6,1])
    plots[n].grid()
    plots[n].axvline(0.6, color='red')
    return days

boyA = generate_baby(15, 0)
boyB = generate_baby(45, 1)
boyC = generate_baby(100, 2)
print('Hospital A:', boyA)
print('Hospital B:', boyB)
print('Hospital C:', boyC)
```
